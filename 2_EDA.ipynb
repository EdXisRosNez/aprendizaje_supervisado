```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/titanic_clean.csv')
```

Distribución de las edades

```python
plt.figure(figsize = (12, 8))

sns.histplot(df, x = 'Age', kde = True)
plt.title('Distribución de Edades')
plt.axvline(df['Age'].mean(), color = 'red')
plt.axvline(df['Age'].median(), color = 'green')
plt.axvline(df['Age'].mode()[0], color = 'black')
plt.show()

# Print the results
print(f"Media: {df['Age'].mean()}")
print(f"Mediana: {df['Age'].median()}")
print(f"Moda: {df['Age'].mode()[0]}")
```

# Matriz de correclación
Esta matriz nos dice qué tan correlacionadas están una variable de la otra

```python
plt.figure(figsize = (12, 8))
sns.heatmap(df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr(), 
            annot=True, 
            cbar=True,  # Activar la barra de colores
            cmap='coolwarm',  # Cambiar la paleta de colores
            cbar_kws={'label': 'Nivel de correlación'})  # Etiqueta para la barra de colores
plt.title('Correlación de variables numéricas')
plt.show()
```

## 5- Agrupa las edades en categorías (Infante, Niño, Adolescente, Joven Adulto, Adulto, Mayor) y crea una gráfica que muestre la tasa de supervivencia por grupo de edad.

```python
infante = df[df['Age'] < 3]
niños = df[(df['Age'] < 13) & (df['Age'] >= 3)]
adolescentes = df[(df['Age'] < 18) & (df['Age'] >= 13)]
joven_adulto = df[(df['Age'] < 30) & (df['Age'] >= 18)]
adulto = df[(df['Age'] < 65) & (df['Age'] >= 30)]
adulto_mayor = df[df['Age'] >= 65]
```

```python
# Calcular la tasa de supervivencia para cada grupo de edad
tasa_infante = infante['Survived'].mean()
tasa_niños = niños['Survived'].mean()
tasa_adolescentes = adolescentes['Survived'].mean()
tasa_jovenadulto = joven_adulto['Survived'].mean()
tasa_adulto = adulto['Survived'].mean()
tasa_adulto_mayor = adulto_mayor['Survived'].mean()
```

```python
# Graficar la tasa de supervivencia por grupo de edades
plt.figure(figsize=(12,8))
grupos_edad = ['Infante', 'Niños', 'Adolescentes', 'Joven Adulto', 'Adulto', 'Adulto Mayor']
tasas_supervivencia = [tasa_infante, tasa_niños, tasa_adolescentes, tasa_jovenadulto, tasa_adulto, tasa_adulto_mayor]

sns.barplot(x=grupos_edad, y=tasas_supervivencia, palette='managua')
```
